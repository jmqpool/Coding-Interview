### Cycle Detection 

A linked list is said to contain a cycle if any node is visited more than once while traversing the list.

Complete the function provided for you in your editor. It has one parameter: a pointer to a Node object named that points to the head of a linked list. Your function must return a boolean denoting whether or not there is a cycle in the list. If there is a cycle, return true; otherwise, return false.



```python
class Node(object):
    def __init__(self, data = None, next_node = None):
        self.data = data
        self.next = next_node

# Move through the list with 2 runners
# runner2 : moves two steps at a time.
# runner1 : moves one step at a time.
# if runner2 reaches the end of the list there is no cycle.
# if runner2 is on the same node as runner1 there is a cycle.

def has_cycle(head):
    
    runner1 = head
    runner2 = head
    
    while (runner2 != None) :
        if (runner2.next == None) or (runner2.next.next == None) :
            return False
        runner1 = runner1.next
        runner2 = runner2.next.next
        if (runner1 == runner2):
            return True
```

### Find Merge Point of Two Lists

Given pointers to the head nodes of linked lists that merge together at some point, find the Node where the two lists merge. It is guaranteed that the two head Nodes will be different, and neither will be NULL. The two separate linked lists that are guaranteed to converge.
Each Node has a data field containing an integer; return the integer data for the Node where the two lists converge.



```python
"""
 Find the node at which both lists merge and return the data of that node.
 head could be None as well for empty list
 Node is defined as
 
 class Node(object):
 
   def __init__(self, data=None, next_node=None):
       self.data = data
       self.next = next_node

 
"""

def FindMergeNode(headA, headB):
    # Determine the lengths of both linked lists.
    nodeA = headA
    nA = 1
    nodeB = headB
    nB = 1  
    while (nodeA.next!=None):
        nodeA = nodeA.next
        nA = nA +1
    while (nodeB.next!=None):
        nodeB = nodeB.next
        nB = nB +1

    # Determine the number first node that list A and B can merge.
    n = min(nA, nB)
    
    # Reset the runners to the heads of the list.
    nodeA = headA
    nodeB = headB
    
    if (nA > n):
        for i in range(nA-n):
            nodeA = nodeA.next
    if (nB > n):
        for i in range(nB-n):
            nodeB = nodeB.next
            
    # The lists with nodeA and nodeB as the head now have the same length.
    # Increment through the nodes until nodeA = nodeB and return the value
    
    for i in range(n):
        if (nodeA == nodeB):
            return(nodeA.data)
        nodeA = nodeA.next
        nodeB = nodeB.next
            
```

### Reverse a doubly linked list
Youâ€™re given the pointer to the head node of a doubly linked list. Reverse the order of the nodes in the list. The head node might be NULL to indicate that the list is empty.

Input Format
You have to complete the Node\* Reverse(Node\* head) method which takes one argument - the head of the doubly linked list. 

Output Format
Change the next and prev pointers of all the nodes so that the direction of the list is reversed. Then return the head node of the reversed list.

```python
"""
 Reverse a doubly linked list
 head could be None as well for empty list
 Node is defined as
 
 class Node(object):
 
   def __init__(self, data=None, next_node=None, prev_node = None):
       self.data = data
       self.next = next_node
       self.prev = prev_node

 return the head node of the updated list 
"""
def Reverse(head):
    # if an empty linked list return the head  
    if(head == None):
        return(head)
    
    # flip the head of the linked list:  the next element should now be None.
    curr_node = head
    curr_node.prev = curr_node.next
    curr_node.next = None
    
    # go through the list incrementally reversing the pointers.
    while(curr_node.prev != None):
        curr_node = curr_node.prev
        next_node = curr_node.prev
        prev_node = curr_node.next
        curr_node.prev = prev_node
        curr_node.next = next_node
        
    # set the head of the reversed list to be the current (last) node  
    head = curr_node
    
    return(head)
```

### Delete duplicate-value nodes from a sorted linked list
You're given the pointer to the head node of a sorted linked list, where the data in the nodes is in ascending order. Delete as few nodes as possible so that the list does not contain any value more than once. The given head pointer may be null indicating that the list is empty.

Output Format
Delete as few nodes as possible to ensure that no two nodes have the same data. Adjust the next pointers to ensure that the remaining nodes form a single sorted linked list. Then return the head of the sorted updated linked list. 

```python
"""
 Delete duplicate nodes
 head could be None as well for empty list
 Node is defined as
 
 class Node(object):
 
   def __init__(self, data=None, next_node=None):
       self.data = data
       self.next = next_node

 return back the head of the linked list in the below method.
"""

def RemoveDuplicates(head):
    #If empty or single element return head
    if( head == None or head.next == None):
        return head
    
    #Trace through the linked list with two nodes, a runner and a follower initialized one node apart
    runner = head.next
    follower = head
    
    while(runner != None):
        #if the runner and follower have the same data values, increment the runner until they are different.   If runner gets to the end of the linked list, then the follower is the last element.
        while(runner.data == follower.data):
            if(runner.next == None):
                follower.next = None
                return(head)
            runner = runner.next
        
        #Update the pointer
        follower.next = runner
        
        #increment down the list
        runner = runner.next
        follower = follower.next
        
    return(head)
```
